```python
from graphics import *
import random


# Randraw by Aditya Sharma
# Code_In_Place Submission


def random_color():
    return color_rgb(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

def random_point(canvas_width, canvas_height):
    return Point(random.randint(0, canvas_width), random.randint(0, canvas_height))

def draw_random_circle(canvas):
    center = random_point(canvas.getWidth(), canvas.getHeight())
    radius = random.randint(10, 100)
    circle = Circle(center, radius)
    circle.setFill(random_color())
    circle.setOutline(random_color())
    circle.setWidth(random.randint(1, 5))
    circle.draw(canvas)

def draw_random_rectangle(canvas):
    p1 = random_point(canvas.getWidth(), canvas.getHeight())
    p2 = random_point(canvas.getWidth(), canvas.getHeight())
    rect = Rectangle(p1, p2)
    rect.setFill(random_color())
    rect.setOutline(random_color())
    rect.setWidth(random.randint(1, 5))
    rect.draw(canvas)

def draw_random_polygon(canvas):
    num_points = random.randint(3, 6)
    points = [random_point(canvas.getWidth(), canvas.getHeight()) for i in range(num_points)]
    poly = Polygon(points)
    poly.setFill(random_color())
    poly.setOutline(random_color())
    poly.setWidth(random.randint(1, 5))
    poly.draw(canvas)

def draw_random_line(canvas):
    p1 = random_point(canvas.getWidth(), canvas.getHeight())
    p2 = random_point(canvas.getWidth(), canvas.getHeight())
    line = Line(p1, p2)
    line.setFill(random_color())
    line.setWidth(random.randint(1, 5))
    line.draw(canvas)

def generate_random_art():
    canvas_width, canvas_height = 800, 600
    canvas = GraphWin("Randraw - Aditya Sharma", canvas_width, canvas_height)
    canvas.setBackground(random_color())

    total_shapes = random.randint(50, 150)

    for i in range(total_shapes):
        shape_type = random.choice(['circle', 'rectangle', 'polygon', 'line'])

        if shape_type == 'circle':
            draw_random_circle(canvas)
        elif shape_type == 'rectangle':
            draw_random_rectangle(canvas)
        elif shape_type == 'polygon':
            draw_random_polygon(canvas)
        elif shape_type == 'line':
            draw_random_line(canvas)

    try:
        # Prompt to click before closing
        message = Text(Point(canvas_width // 2, 20), "Click anywhere to exit, Thank you")
        message.setTextColor("white")
        message.draw(canvas)
        canvas.getMouse()
    except GraphicsError:
        pass  # Ignore if the window was closed manually

    canvas.close()

# Run the program
generate_random_art()
```
